```python
import pandas
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt


drug = pandas.read_csv("drug200.csv")
```

```python
d = {'HIGH': 0, 'NORMAL': 1, 'LOW': 2}
drug['BP'] = drug['BP'].map(d)

d = {'HIGH': 0, 'NORMAL': 1, 'LOW':2}
drug['Cholesterol'] = drug['Cholesterol'].map(d)

d = {'F': 0, 'M': 1}
drug['Sex'] = drug['Sex'].map(d)

d = {'drugA': 0, 'drugB': 1, 'drugC': 2, 'drugX': 3, 'drugY': 4}
drug['Drug'] = drug['Drug'].map(d)

features = ['Age', 'Sex', 'BP', 'Cholesterol']


X = drug[features]
y = drug['Drug']


dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)


tree.plot_tree(dtree, feature_names=features)
```

```python
print(dtree.predict([[40, 14, 8, 3]]))


print("[1] means 'drugA'")
print("[0] means 'drugB'")
print("[2] means 'drugC'")
print("[3] means 'drugX'")
print("[4] means 'drugY'")
```
